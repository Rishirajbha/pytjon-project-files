# Which Month Maximum Number Fire were Repoted

```python
df1=df.groupby('month') ['number'].sum().reset_index().sort_values(by='number',ascending=False)
df1
```

```python
# In the month of july max no of fire were repoted.
```

```python
sys.barplot(x='month',y='number',data=df1)
plt.xticks(rotation=65)
plt.title('Fire Incedent Repoted on Monthly Basis')
plt.show()
```

# In which Year Max Number of Forest Fire were Repoted

```python
df.columns
```

```python
df2=df.groupby('year') ['number'].sum().reset_index().sort_values(by='number',ascending=False)
df2
```

```python
plt.figure(figsize=(16,5))
sys.barplot(x='year',y='number',data=df2)
plt.title('Fire Incedent Repoted on Yearly Basis')
plt.show()
```

# In which State Max Number of Forest Fire were Repoted

```python
df3=df.groupby('state') ['number'].sum().reset_index().sort_values(by='number',ascending=False)
df3
```

```python
plt.figure(figsize=(16,5))
sys.barplot(x='state',y='number',data=df3)
plt.xticks(rotation=75)
plt.title('Fire Incedent Repoted on Yearly Basis',size=25,color='Red')
plt.show()
```

# Total No of Fires were Repoted in 2015 & Visualize Data Based on Each Month

```python
df.columns
```

```python
fire_repoted=df[df['year']==2015].groupby('month') ['number'].sum().reset_index()
fire_repoted
```

```python
plt.figure(figsize=(16,5))
sys.barplot(x='month',y='number',data=fire_repoted)
plt.xticks(rotation=65)
plt.title("Tot.Number of Fire Repoted in 2015,Based on Each Month",size=25,color='Red')
plt.show()
```

# AVerage No. of Fires were Repoted from Highest to lowest (State-Wise)

```python
df.columns
```

```python
w=df.groupby('state')['number'].mean().sort_values(ascending=False).reset_index()
w
```

```python
plt.figure(figsize=(16,5))
sys.barplot(x='state',y='number',data=w)
plt.xticks(rotation=65)
plt.title("Average No. of Fire were Repoted in State Wise",size=25,color='Red')
plt.show()
```
