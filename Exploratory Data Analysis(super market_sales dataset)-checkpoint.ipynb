## Display Daily Sales by Day of the Week

```python
df.columns
```

```python
day_map={
         0:'Mon',
         1:'Tue',
         2:'Wed',
         3:'Thus',
         4:'Fri',
         5:'Sat',
         6:'Sun'}
```

```python
df['Day_of_week']=df['Date'].dt.dayofweek.map(day_map)
```

```python
df.head(1)
```

```python
df['Day_of_week'].value_counts().plot(kind='bar')
```
