# 15. Another column for replace value of 'income' with (0,1)

```python
df['income'].unique()
```

```python
df['salary']=df['income']
```

```python
sys.countplot('salary',data=df)
```

```python
def salary_value (sal):
    if sal== '<=50K':
        return 0
    else:                   
        return 1
    print(salary_value)
```

```python
df['encoded_sal']=df['salary'].apply(salary_value) # maked another column and replace the salary with (0,1) value
```
